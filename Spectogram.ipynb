```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sound Effect 
# by Kitten Strike: https://pixabay.com/users/kittenstrike1-35556891/?utm_source=link-attribution&utm_medium=referral&utm_campaign=music&utm_content=158832 
# from Pixabay: https://pixabay.com/sound-effects//?utm_source=link-attribution&utm_medium=referral&utm_campaign=music&utm_content=158832
data = 'quothello-therequot-158832.mp3'
```

```python
# Code from https://medium.com/analytics-vidhya/understanding-the-mel-spectrogram-fca2afa2ce53
```

```python
y, sr = librosa.load(data)
plt.figure(figsize=(9,3))
plt.plot(y)
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.subplots_adjust(bottom=0.2)
plt.savefig('.\summary\plots\signal.png')
```

"there": /ðɛr/ or /ðɛər/

```python
n_fft = 2048
ft = np.abs(librosa.stft(y[:n_fft], hop_length = n_fft+1))
plt.figure(figsize=(9,3))
plt.plot(ft)
plt.xlabel('Frequency Bin')
plt.ylabel('Amplitude')
plt.subplots_adjust(bottom=0.2)
plt.savefig('.\summary\plots\spec.png')
```

```python
spec = np.abs(librosa.stft(y, hop_length=512))
spec = librosa.amplitude_to_db(spec, ref=np.max)
librosa.display.specshow(spec, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.savefig('.\summary\plots\spectogram.png')
```

```python
mel_spect = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=1024)
mel_spect = librosa.power_to_db(mel_spect, ref=np.max)
librosa.display.specshow(mel_spect, y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.savefig('.\summary\plots\mel_spectogram.png')
```
